```python
#importing libraries
import numpy as np
```

```python
#ASCII conversion
d={chr(i):i for i in range(255)} #char to askii
c={ i:chr(i) for i in range(255)} #askii to char conversion
```

```python
#message and encryption key
key="key"
text="hello"
```

```python
#random pixel value image of 10 x 10 size with 3 color channels
x=np.random.randint(0,255,(10,10,3),dtype=np.uint8)
```

```python
x
```

```python
#encrypt part
# Make a copy of the image to embed text
x_enc = x.copy()
n = 0  # row
m = 0  # column
z = 0  # color channel
l = len(text)
kl = 0  # index for key

for i in range(l):
    # XOR the ASCII values of text and key characters
    char_val = d[text[i]] ^ d[key[kl]]

    for bit_pos in range(8):
        # Get the bit at position bit_pos (MSB to LSB) ie take the lsb from 8 bits given to us
        bit = (char_val >> (7 - bit_pos)) & 1

        orig_val = x_enc[n, m, z]

        # Embed the bit in the LSB of the color value
        x_enc[n, m, z] = (orig_val & 0b11111110) | bit

        z = (z + 1) % 3  # move to next channel (BGR)

        # Move to next pixel if needed
        if z == 0:
            m += 1
            if m == x_enc.shape[1]:  # if end of row, go to next row
                m = 0
                n += 1

    kl = (kl + 1) % len(key)  # cycle key index
```

```python
#decrypt
# Initialize indices and variables
n, m, z = 0, 0, 0
kl = 0
l=len(text)
decrypt = ""

# Loop to read one character (8 bits)
for i in range(l):  # You can increase this range to decrypt more characters
    val = 0

    # Read 8 bits to form one character
    for bit_pos in range(8):
        # Extract the least significant bit from x_enc[n][m][z]
        bit = x_enc[n, m, z] & 1

        # Shift the current val and add the new bit
        val = (val << 1) | bit

        # Print current bit and its position
        print(f"Reading bit {bit} from ({n},{m},{z})")

        # Move to the next z index, wrap around using modulo 3
        z = (z + 1) % 3

        # If z has looped back to 0, move to the next m
        if z == 0:
            m += 1

            # If m exceeds the width, reset m and move to next n (row)
            if m == x_enc.shape[1]:
                m = 0
                n += 1

    # Decrypt the character using key and XOR logic
    orig_char = c[val ^ d[key[kl]]]
    decrypt += orig_char

    print(f"Decrypted bytes: {val} XOR {d[key[kl]]}= {val ^ d[key[kl]]} -> '{orig_char}'")

    # Move to the next character in key (cyclically)
    kl = (kl + 1) % len(key)

# Print the final decrypted text
print("Decrypt text", decrypt)
```
